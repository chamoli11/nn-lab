```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x= np.linspace(0,10, 1000)
plt.plot(x, np.sin(x), label= 'sin(X)', color='g')
plt.plot(x, np.cos(x), label= 'cos(X)', color= 'r')
plt.xlabel('X')
plt.ylabel('SIN X')
plt.legend(loc = 'lower right')
```

```python
x= np.linspace(0, 10, 1000)
plot= plt.figure(figsize=(10,6))
figu, ax= plt.subplots(2,2)
ax[0,0].plot(x, np.sin(x), color= 'r', linestyle= 'dashed')
ax[0,1].plot(x, np.cos(x), color= 'g', linestyle= 'dotted')
```

```python
cs=[20, 45, 8, 70, 34]
ec=[75,30 ,12, 88, 28]
barw=0.25
bar1= np.arange(len(cs))
bar2= [x+barw for x in bar1]

plt.bar(bar1,cs, width= barw, label= 'cs')
plt.bar(bar2, ec, width=barw, label= 'ec')
plt.legend()
```
